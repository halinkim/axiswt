```python
import numpy as np
import matplotlib.pyplot as plt

# original data
data = np.concatenate((np.sin(2 * np.pi * np.linspace(0, 10, 256)) + np.sin(2 * np.pi * np.linspace(0, 5, 256)),
                        2 * np.sin(2 * np.pi * np.linspace(0, 10, 256)),
                        4 * np.sin(2 * np.pi * np.linspace(0, 5, 256)),
                       3 * np.sin(2 * np.pi * np.linspace(0, 3, 255))))

plt.plot(data)
```

```python
import axiswt # newmade
import pywt # original
```

```python
def test(method, wavelets='sym5'):
    low, high = method(data, wavelets, 'smooth')
    low2, high2 = method(low, wavelets, 'smooth')
    
    plt.figure(figsize=(8, 2), dpi=300)
    plt.subplot(2, 2, 1)
    plt.plot(low)
    plt.subplot(2, 2, 2)
    plt.plot(high)
    plt.subplot(2, 2, 3)
    plt.plot(low2)
    plt.subplot(2, 2, 4)
    plt.plot(high2)
    plt.tight_layout()
```

```python
test(axiswt.dwt_forward, 'sym5')
```

```python
test(pywt.dwt, 'sym5')
```

```python
test(axiswt.dwt_forward, 'haar')
```

```python
test(pywt.dwt, 'haar')
```
